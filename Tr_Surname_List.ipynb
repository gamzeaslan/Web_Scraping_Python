```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://forebears.io/turkey/surnames"
```

```python
header={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"}
```

```python
response=requests.get(url,headers=header)
```

```python
print(response)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
find=soup.find("div",{"class":"table-responsive distribution-holder","id":"names-2014"}).find_all("tbody")
```

```python
surname_list=[]
```

```python

for item in find:
  names= item.find_all("td",{"class":"sur"})
  for name in names:
    surname_list.append(name.find("a").text)
```

```python
for item in surname_list:
  print(item)
```

```python
len(surname_list)
```

```python
import pandas as pd
```

```python
dic={"surname":surname_list}
```

```python
df=pd.DataFrame(dic)
```
